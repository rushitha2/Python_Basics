### Arguments

```python
# positional arguments
def add_two(a,b):
    print(f"a is {a}")
    print(f"b is {b}")
    print(a+b)

add_two(2,3)
add_two(3,2)
```

```python
# keyword arguments
def info (name,age):
    print(f"i am {name} and my age is {age}")

info(13,'jin')
```

```python
info(age=30,name='jin')
```

```python
# default arguments
def info (name,age=15):
    print(f"i am {name} and my age is {age}")
```

```python
info('jin')
info('jk',25)
```

```python
# arbitary arguments
def total (*args):
    print(args)
    return sum(args)

print(total(1,2,3,4))
print(total(1,2))
```

```python
def print_all (*args):
    for i in args:
        print(i)
print_all('a','b','c')
```

```python
# arbitary keyword arguments
def info(**kwargs):
    print(kwargs)
    for key,value in kwargs.items():
        print(f"{key} is {value}")

info(name='rushi',age=44,color='red')
```

```python
# mix of all arguments
def profile(name,email,*args,**kwargs):
    print(f"name is {name}")
    print(f"email is {email}")
    for i in args:
        print(i)
    for key,value in kwargs.items():
        print(f"{key} is {value}")
    
```

```python
profile(
    'rushi','rushitha03@gmail.com',
    'loves bts','loved by all',
    age=3,height=164, branch='cse'
)
```

```python
# filter()
"""
filter() takes two arguments (function,iterable)
returns true or false which is boolean
we can only check a condition but not modify the items
"""

def positive(n):
    if n>0:
        return n
```

```python
# map()
"""
map() takes two arguments (function,iterable)
we can  modify the items
"""
def square(n):
    return n**2
```

```python
a = [1,2,3]
sq = list(map(square,a))
print(sq)
```

```python
from functools import reduce

def multiply(x,y):
    return x*y
a =[1,2,2,2]
b = reduce(multiply,a)
print(b)
```

```python
# isinstance() -- checks the data structure and returns true if matches else false
a = [1,2,3]

print(isinstance(a,list))
print(isinstance(a,tuple))
```

```python
# Recursive Functions
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)
print(factorial(5))
```

```python
def rev_string (s):
    if len(s) == 1:
        return s[0]
    else:
        return rev_string(s[1:]) + s[0]

print(rev_string('bye'))
```
